```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
%matplotlib inline
```

```python
## Read data files .asc
data = np.loadtxt("data1.asc")
#print(data)

df = pd.DataFrame(data, columns=['spatial1','spatial2','time','physical','intensity'])
#df

## Re-order the data based on the 'time' variable/dimension
dfo = df.sort_values('time', axis=0, ascending=True) 
dfo
```

```python
# Show basic information
dfo.describe()
```

```python
# Import plotly.express as px
from mpl_toolkits import mplot3d

# Present the physical property using the size of points
# Put the other three variable in x, y, z axises 

# Put the time variable aside
# We can change the time option 
# to see the data at different time values
# Use 'interact' funtion to show the data

def plottime(time):
    dfo.sort_values(['physical'], inplace=True, ascending=False)
    Time = df.time == time
    area = df[Time].physical - 824 
    # since the values of 'physical' range from about 824 to 835,
    # the deduction of 824 helps to better visualize the variation
    
    fig = plt.figure(figsize = (9,9))
    #ax = fig.add_subplot(111, projection = '3d')
    ax = plt.axes(projection = '3d')
    
    ## Input those variable values
    x_vals = df[Time].spatial1 
    y_vals = df[Time].spatial2
    z_vals = df[Time].intensity
    
    ax.set_xlabel('spatial 1')
    ax.set_ylabel('spatial 2')
    ax.set_zlabel('intensity')
    
    ## Plot the values
    ax.scatter3D(x_vals, y_vals, z_vals, 
                 s = area, linewidth = 0.7, 
                 edgecolors = 'g');

    
interact(plottime, time=df['time'])   
```
